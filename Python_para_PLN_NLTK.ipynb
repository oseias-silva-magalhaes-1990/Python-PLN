```python
infile = open("/content/drive/MyDrive/python_pln_arq/corpus_teste.txt", 'r')
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w+')#Regex pega apenas palavras e rejeita pontuações e números
tokens = tokenizer.tokenize(infile.read())
print(tokens)

nova_lista = []
for token in tokens:
  nova_lista.append(token.lower())

frequencia = nltk.FreqDist(nova_lista)
print(frequencia.most_common(10))#Pega a frequência
infile.close()
```

```python
stopwords = nltk.corpus.stopwords.words('portuguese')
infile = open("/content/drive/MyDrive/python_pln_arq/corpus_teste.txt", 'r')
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w+')#Regex pega apenas palavras e rejeita pontuações e números
tokens = tokenizer.tokenize(infile.read())
#print(tokens)

nova_lista = []
for token in tokens:
  if token.lower() not in stopwords:#Se não estão na lista de stopwords
    nova_lista.append(token.lower())

frequencia = nltk.FreqDist(nova_lista)
print(frequencia.most_common(10))#Pega a frequência
infile.close()
```

```python
from nltk.corpus import mac_morpho
from nltk.tag import UnigramTagger
from nltk.tag import DefaultTagger

infile = open("/content/drive/MyDrive/python_pln_arq/corpus_teste.txt", 'r').read()

tokens = nltk.word_tokenize(infile)

etiquet_padrao = DefaultTagger("N")
sentencas_treino = mac_morpho.tagged_sents()
etiquetador = UnigramTagger(sentencas_treino, backoff = etiquet_padrao)

etiquetado = etiquetador.tag(tokens)
#print(etiquetado)
```

```python
from nltk.chunk import RegexpParser

pattern = 'NP: {<NPROP><NPROP> | <N><N>}'
analise_gramatical = RegexpParser(pattern)

arvore = analise_gramatical.parse(etiquetado)
print(arvore)
```
